```python
import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings("ignore")
```

## We don`t have any header so header = None with na values represented as ?

```python
df = pd.read_csv("imports-85.data",header=None,na_values="?")
df.head()
```

## Intially we don\`t have header title on our dataset , so adding headers on our data 

```python
headers = ["symboling","normalized_losses","make","fuel_type","aspiration",
           "num_doors","body_style","drive_wheels","engine_location",
           "wheel_base","length","width","height","curb_weight",
           "engine_type","num_cylinders","engine_size","fuel_system",
           "bore","stroke","compression_ratio","horsepower","peak_rpm",
           "city_mpg","highway_mpg","price"]
```

```python
df.columns = headers
```

```python
df.head()
```

## Dealing with Missing Data

```python
df.info()  #Total 205 entries must be there , but some of them are null ; which are referred here by 164 non null entries
```

## To drop Null values in dataset

```python
df_temp = df.dropna()  
```

```python
df_temp.info()
```

We lost 205-159 = 46 row entries inorder to remove null values ; but we are missing a lot of important data this way !!

## Filling Null data with some details as per our need

```python
df["normalized_losses"].fillna(df.normalized_losses.mean(),inplace=True)  # To replace the Null values with the mean of that column
```

```python
df.head()
```

## Filling missing values for categorical values ( The values whose average we can\'t find to replace the null values )

```python
from scipy.stats import mode
mode(df["num_doors"][(df["num_doors"]).notnull()]) #Replacing the null values with Mode of the column values , here mode is four as there are more four gated cars
```

```python
df.info()  #Nullable values reduced as we tried filling NA values with mean,mode etc.
```

## Handling Categorical Data ( The data which is not numeric and are mostly like English words like four,two etc.)

```python
# We will map the catergorical data with some ordinal features like we replace "four" with 1 and "two" with 0
Door_mapping = {"four":1,"two":0}
df["num_doors"] = df["num_doors"].map(Door_mapping)
```

```python
df.head(5)
```
