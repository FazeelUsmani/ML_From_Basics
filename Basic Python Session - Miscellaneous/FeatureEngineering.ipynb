# We will be using Boston Housing Price Dataset for this notebook

```python
import pandas as pd  
df = pd.read_csv("housing.csv")
df.head() 
```

## Highlighting the important characteristics / parameters in the dataset

```python
import warnings 
warnings.filterwarnings('ignore') #To ignore the warnings by Jupyter notebook while running the code

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid" ,context="notebook") 
cols = ["LSTAT","RM","MEDV"]
sns.pairplot(df[cols],size=2.5)
plt.tight_layout()
plt.show() 
```

# Correlation Matrix

The correlation matrix is a square matrix that contains the Pearson product-moment correlation coefficients ( often abbreviated as Pearson's r ), which measure the linear dependence between pairs of features. The correlation coefficients are bounded to the range -1 and 1. Two features have a perfect positive correlation if r = 1 , no correlation if r = 0 and a perfect negative correlation if r = -1 respectively, As mentioned previously, Pearson's correlation coefficient can simply be calculated as the covariance between two features x and y(numerator) divided by the product of their standard deviations (Denominator)

```python
import numpy as np
cm = np.corrcoef(df[cols].values.T) 
sns.set(font_scale = 1.5)
hm = sns.heatmap(cm,
                 cbar = True,
                 annot = True,
                 square = True,
                 fmt = '.2f',
                 annot_kws = {"size":15},
                 yticklabels=cols,
                 xticklabels=cols) 
plt.show()                 
```

- Value close to 1 : Has a strong relationship with the target column
- Value close to 0 : Not so strong relationship ( A positive CoRelationship )
- Value close to -1 : Has a strong relationship ( A Negative CoRelationship )
